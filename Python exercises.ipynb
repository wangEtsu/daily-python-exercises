```python
# Question 6
# Level 2

# Question:
# Write a program that calculates and prints the value according to the given formula:
# Q = Square root of [(2 * C * D)/H]
# Following are the fixed values of C and H:
# C is 50. H is 30.
# D is the variable whose values should be input to your program in a comma-separated sequence.
# Example
# Let us assume the following comma separated input sequence is given to the program:
# 100,150,180
# The output of the program should be:
# 18,22,24

# My Solution
import math
d_seq = input()
d_list = d_seq.split(',')
def simple_calculator(seq):
    output_list = []
    c = 50
    h = 30
    for d in seq:
        d = int(d)
        q = math.sqrt((2*c*d)/h)
        q = round(q)
        output_list.append(q)
    return output_list
simple_calculator(d_list)
```

```python
# Question 7
# Level 2

# Question:
# Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The element value in the i-th row and j-th column of the array should be i*j.
# Note: i=0,1.., X-1; j=0,1,¡­Y-1.
# Example
# Suppose the following inputs are given to the program:
# 3,5
# Then, the output of the program should be:
# [[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]] 

# My Solution
input_seq = input()
ij_list = input_seq.split(',')
i = int(ij_list[0])
j = int(ij_list[1])
result_list = [[0 for j in range(j)] for i in range(i)]
for row in range(i):
    # print("i=", row)
    for col in range(j):
        # print("j=", col) 
        result_list[row][col] = row*col 
result_list
```

```python
# Question 17
# Level 2

# Question:
# Write a program that computes the net amount of a bank account based a transaction log from console input. The transaction log format is shown as following:
# D 100
# W 200

# D means deposit while W means withdrawal.
# Suppose the following input is supplied to the program:
# D 300
# D 300
# W 200
# D 100
# Then, the output should be:
# 500

# My Solution
netAmount = 0
while True:
    transaction = input()
    if not transaction:
        break
    else:
        op = transaction.split(' ')[0]
        amount = int(transaction.split(' ')[1])
        if op == 'D':
            netAmount = netAmount + amount
        elif op == 'W':
            netAmount = netAmount - amount
print(netAmount)
    
```

```python
# Question 21
# Level 3

# Question£º
# A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps. The trace of robot movement is shown as the following:
# UP 5
# DOWN 3
# LEFT 3
# RIGHT 2
# ¡­
# The numbers after the direction are steps. Please write a program to compute the distance from current position after a sequence of movement and original point. If the distance is a float, then just print the nearest integer.
# Example:
# If the following tuples are given as input to the program:
# UP 5
# DOWN 3
# LEFT 3
# RIGHT 2
# Then, the output of the program should be:
# 2

# My Solution
import numpy as np
import math
pos = (0, 0)
while True:   
    move = input()
    if not move:
        break
    direction = move.split(' ')[0]
    distance = int(move.split(' ')[1])
    if direction == "UP":
        pos = np.add(pos,(0, distance))
    elif direction == "DOWN":
        pos = np.add(pos,(0, -distance))
    elif direction == "LEFT":
        pos = np.add(pos,(distance, 0))
    elif direction == "RIGHT":
        pos = np.add(pos,(-distance, 0))
    distance_from_start = math.sqrt(pos[0]**2 + pos[1]**2)
    print(distance_from_start)
```

### Pass

```python
# Question 104
# Write a Python function to check whether a number is perfect or not. Go to the editor
# According to Wikipedia : In number theory, a perfect number is a positive integer that is equal to the sum of its proper positive divisors, that is, the sum of its positive divisors excluding the number itself (also known as its aliquot sum). Equivalently, a perfect number is a number that is half the sum of all of its positive divisors (including itself).
# Example : The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6. Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6. The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128. 

# My Solution
def perfect_checker(num):
    divisors = [i for i in range(1, num) if num%i==0]
    total = 0
    for divisor in divisors:
        total += divisor
    if total == num:
        return True
    else:
        return False
perfect_checker(28)
```

```python
# Question 106
# Write a Python function that prints out the first n rows of Pascal's triangle. Go to the editor
# Note : Pascal's triangle is an arithmetic and geometric figure first imagined by Blaise Pascal.

# My Solution
def pascal_printer(n):
    # Used to store the whole pascal triangle
    triangle = []
    for i in range(n):
        # used to store each row, start with 1
        row = [1]      
        for j in range(i):
            # In case i = 0, or its the first row, pass
            if i == 0:
                pass
            # Else, append 1 as the last element, and append elements according to previous rows
            else:
                if j != i-1:
                    row.append(triangle[i-1][j] + triangle[i-1][j+1])
                elif j == i-1:
                    row.append(1)
        triangle.append(row)
    print(triangle)
pascal_printer(10)
```

### Decorator
Modifying an function dynamically without changing its content permanently

```python
# defining a decorator 
def hello_decorator(func): 
  
    # inner1 is a Wrapper function in  
    # which the argument is called 
      
    # wrapper function can access the outer local 
    # functions like in this case "func" 
    def wrapper_function(): 
        print("Hello, this is before function execution") 
  
        # calling the actual function now 
        # inside the wrapper function. 
        return func()         
    return wrapper_function 

# use @ to pass display in as an argument of hellp_decorator  
@hello_decorator
def display():
    print("Hi, This is display function")

# check the result
display()
```

```python
# Question 110
# Write a Python program to make a chain of function decorators (bold, italic, underline etc.) in Python

# Their Solution
def make_bold(fn):
    def wrapped():
        return "<b>" + fn() + "</b>"
    return wrapped

def make_italic(fn):
    def wrapped():
        return "<i>" + fn() + "</i>"
    return wrapped

def make_underline(fn):
    def wrapped():
        return "<u>" + fn() + "</u>"
    return wrapped
@make_bold
@make_italic
@make_underline
def hello():
    return "hello world"
print(hello()) ## returns "<b><i><u>hello world</u></i></b>"
```
